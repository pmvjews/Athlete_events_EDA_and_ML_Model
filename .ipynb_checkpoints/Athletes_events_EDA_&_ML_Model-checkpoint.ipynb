# Feature engineering

```python
df['Weight'] = round(df['Weight'].fillna(df.groupby(['Team'])['Weight'].transform('mean')), 2)
df['Weight'] = round(df['Weight'].fillna(df.groupby(['Age'])['Weight'].transform('mean')), 2)
df['Weight'] = round(df['Weight'].fillna(df.groupby(['Sex'])['Weight'].transform('mean')), 2)

df['Height'] = round(df['Height'].fillna(df.groupby(['Team'])['Height'].transform('mean')), 2)
df['Height'] = round(df['Height'].fillna(df.groupby(['Age'])['Height'].transform('mean')), 2)
df['Height'] = round(df['Height'].fillna(df.groupby(['Sex'])['Height'].transform('mean')), 2)

df = df[df['Age'].notna()]
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
filter1 = df["Year"] >= 1900
filter2 = df["Year"] <= 1999

df.where(filter1 & filter2, inplace=True)
df = df[df['Sport'] == "Athletics"]

df.head()
```

```python
df.reset_index(inplace=True)
df = df[['Sex', 'Age', 'Height', 'Weight', 'Year',  'Medal']]

df.head()
```

```python
df.isnull().sum()
```

```python
df['Medal'] = [1 if i == 'Bronze' or i == 'Silver' or i == 'Gold' else 0 for i in df['Medal']]
df['Sex'] = [1 if i == 'M' else 0 for i in df['Sex']]
```

```python
df['Medal'].unique()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Age'] = df['Age'] / 52
df['Height'] = df['Height'] / 208
df['Weight'] = df['Weight'] / 150
df['Year'] = df['Year'] / 1996
```

```python
df.describe()
```

# Building a model

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.loc[:, ['Sex', 'Age', 'Height', 'Weight', 'Year']].values
y = df.loc[:, ['Medal']].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
# model = tf.keras.Sequential()
# model.add(tf.keras.layers.Dense(128, input_shape=(5,),  activation=keras.layers.LeakyReLU(alpha=0.2))),
# model.add(tf.keras.layers.Dense(64, activation=keras.layers.LeakyReLU(alpha=0.2))),
# model.add(tf.keras.layers.Dense(2, activation='softmax')),
# model.summary()

model = tf.keras.Sequential([
    tf.keras.layers.Dense(20612, input_shape=(5,),  activation=keras.layers.LeakyReLU(alpha=0.2)),
    tf.keras.layers.Dense(64, activation=keras.layers.LeakyReLU(alpha=0.2)),
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(2, activation='softmax')
])
model.summary()


# model = tf.keras.Sequential()
# model.add(tf.keras.layers.Dense(64, activation='relu', input_shape=(5,)))
# model.add(tf.keras.layers.Dense(256))
# model.add(tf.keras.layers.Dense(2))
# model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(
    learning_rate=0.3
),
             loss='binary_crossentropy',
             metrics=['binary_crossentropy', 'accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10)
```
